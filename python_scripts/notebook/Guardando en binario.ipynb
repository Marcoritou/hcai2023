```python
import numpy as np

rand1 = np.loadtxt('datos.txt', usecols=[1], unpack= True)
```

```python
rand1.shape
```

```python
nuevo_rand1 = rand1.copy()
```

```python
import struct


datos_binarios = struct.pack('f'*rand1.shape[0],*tuple(nuevo_rand1))
```

```python
with open('datos_binario.araucaria', 'wb') as f:
    f.write(datos_binarios)
```

```python
with open('datos_binario.araucaria', 'rb') as f:
    lectura_bin_data = f.read()
```

```python
struct.unpack(''*rand1.shape[0], lectura_bin_data)
```
