```python
hdf_file = h5py.File('datos.h5',mode='r+')

hdf_file.attrs['Description'] = 'Archivo con datos generados con una distribucion normal, a traves de numpy random'

otros_datos = np.random.exponential(size=10**3)

#hdf_file['Datos exponenciales/Set1'] = otros_datos

hdf_file['Datos exponenciales/Set1'].attrs['Description'] = 'Datos generados mediante una distribucion exponencial con numpy'
```
